```python
import numpy as np
import matplotlib.pyplot as plt

# Load the spike vector and plain vector from .npy files
spike_vector = np.load('spikes_vector_1000.npy')
plain_vector = np.load('plain_vector_1000.npy')

# Assuming each vector is a segment with the same length
num_spikes = spike_vector.shape[0]
spike_length = spike_vector.shape[1]  # Number of samples per spike
num_plain = plain_vector.shape[0]

# Create a time axis for plotting (adjust as necessary)
time = np.arange(spike_length)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot spikes
for i in range(20):
    plt.plot(time, spike_vector[i], alpha=0.5, label='Spike' if i == 0 else "")

# Plot plain vector
# for j in range(20):
#     plt.plot(time, plain_vector[j], alpha=0.5, color='orange', label='Plain' if j == 0 else "")

# Customize the plot
plt.title('Comparison of Spike Vector and Plain Vector')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.xlim(0, 200)
plt.ylim(-300,300)
plt.grid()
plt.legend()
plt.show()
```
